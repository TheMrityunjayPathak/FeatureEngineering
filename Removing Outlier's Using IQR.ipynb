**Interquartile Range**

**IQR = Interquartile Range**

**Q3 = 3rd quartile or 75th percentile**

**Q1 = 1st quartile or 25th percentile**

**Importing Pandas Library**

```python
import pandas as pd
```

**Reading CSV File**

```python
df = pd.read_csv("heights.csv")
df
```

**Using the describe() Function to get Statistical Information**

```python
df.height.describe()
```

**Calculating Q1 and Q3**

```python
Q1 = df.height.quantile(0.25)
Q3 = df.height.quantile(0.75)
Q1,Q3
```

**Calculating IQR**

```python
IQR = Q3 - Q1
IQR
```

**Calculating Upper Limit and Lower Limit**

```python
upper_limit = Q3 + 1.5*IQR
lower_limit = Q1 - 1.5*IQR
upper_limit,lower_limit
```

**Detecting Outlier's in Data**

```python
df[(df.height<lower_limit) | (df.height>upper_limit)]
```

**Removing Outlier's from Data**

```python
df_without_outlier = df[(df.height>lower_limit) & (df.height<upper_limit)]
df_without_outlier
```
